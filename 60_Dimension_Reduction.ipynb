# Principal Component Analysis (PCA)

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)

#base data

data1 = np.random.normal(loc=10,scale=2,size=100)

data2 = data1*0.8 + np.random.normal(loc=0,scale=1,size=100)

data3 = data1 * 0.5 + np.random.normal(loc=0, scale = 1.5, size = 100)
```

```python
#use vertical stack to multiple data into a single dataset

X = np.vstack((data1,data2,data3)).transpose()
```

```python
print(X)
```

```python
scaler = StandardScaler()
X_scaled=scaler.fit_transform(X)

print(X_scaled)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,8))

ax1=fig.add_subplot(1,2,1,projection='3d')
ax1.scatter(X[:,0],X[:,1],X[:,2], c='skyblue',marker='o',alpha=0.7)
```

```python
#Apply PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

# What is Explained Variance Ratio
When PCA reduced the dimensionality of your data, it essentially finds new dimensions (called principal components) that capture the most variances in the original data

```python
variance_ratio = pca.explained_variance_ratio_

print(f"Explained Variance Ratio of components {variance_ratio.sum():.2f}")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,8))

ax1=fig.add_subplot(1,2,1,)
ax1.scatter(X_pca[:,0],X_pca[:,1], c='red',marker='x',alpha=0.7)
```

```python
# Iris data set

from sklearn.datasets import load_iris
```

```python
iris = load_iris()

data_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
data_df.head()
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(data_df)
```

```python
pca = PCA(n_components=2)

pca_data = pca.fit_transform(scaled_data)
```

```python
pca_data
```

```python
pca_df = pd.DataFrame(data=pca_data, columns=['PC1','PC2'])
```

```python
pca_df
```

```python
variance_ratio = pca.explained_variance_ratio_

print(f"Explained Variance Ratio of components {variance_ratio.sum():.2f}")
```

```python
#Applying K Means clustering

kmeans = KMeans(n_clusters=3,random_state = 42)

clusters = kmeans.fit_predict(pca_df)
```

```python
pca_df['clusters'] = clusters
pca_df
```

```python
plt.figure(figsize=(6,8))
plt.scatter(pca_df['PC1'],pca_df['PC2'],c=pca_df['clusters'],cmap='viridis',s=50,alpha=0.8)
#
```
