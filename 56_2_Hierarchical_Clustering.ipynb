```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
data = {
    'Name': ['Alan','Lisa','Joe','Max','Caro'],
    'Social Media': [2,5,5,1,4],
    'Gym': [3,2,3,4,5]
}
df = pd.DataFrame(data)
df
```

```python
X = df[['Social Media', 'Gym']]

# Standardization --> will bring your value to an equal scale
#e.g. height (140-190cm) / Age (14-20)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#Performing Hierarchical Clustering

linked = linkage(X_scaled, method='ward')
linked

#Plot dendrogram

plt.figure(figsize=(6, 4))
dendrogram(linked,labels=df['Name'].values,distance_sort='ascending',show_leaf_counts=True)
```

```python
# Exercise

data = {
    'Customer': ['Amy', 'Brian', 'Clara', 'David', 'Eva'],
    'Visits per week': [5, 1, 4, 2, 5],
    'Avg Spend': [8, 20, 9, 25, 7]
}
df = pd.DataFrame(data)
df
```

```python
X = df[['Visits per week','Avg Spend']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#Performing Hierarchical Clustering

linked = linkage(X_scaled, method='ward')
linked

#Plot dendrogram

plt.figure(figsize=(6, 4))
dendrogram(linked,labels=df['Customer'].values,distance_sort='ascending',show_leaf_counts=True)
```
