# K Means Clustering

```python
iris_df = iris_df.drop('target', axis = 1)
```

```python
iris_df
```

```python
X = iris.data
iris_df
```

```python
X
```

```python
# choose the number of clusters - K
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X[:2])
print(X_scaled[:2])
```

```python
df_scaled = pd.DataFrame(X_scaled, columns=iris.feature_names)
df_scaled.head()
```

```python
wcss = []
k_values = range(1,11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
#
```

```python
print(wcss)
```

```python
plt.figure(figsize=(12,8))
plt.plot(k_values, wcss, 'bx-',markersize = 8)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
#Implementing KMeans with n_cluster = 3

k =3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)
```

```python
#get the clusters and the centroids

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(labels)
print(centroids)
```

```python
#Plot the cluster with centroids as well

for i in range(k):
    plt.scatter(X_scaled[labels==i,0], X_scaled[labels==i,1], label = f'Cluster {i}')

plt.scatter(centroids[:,0], centroids[:,1], s=200, c='red')
plt.legend()
plt.show()
```
